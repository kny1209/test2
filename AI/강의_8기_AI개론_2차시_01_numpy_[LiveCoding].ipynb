##응용 예시

```python
# 두 변수를 비교해서 일치하는 개수로 정확도 계산


# 2개의 넘파이 배열 준비


# 정답
yt = np.array([1,1,0,1,0,1,1,0,1,0])


# 예측
yp = np.array([1,1,0,1,0,1,0,1,1,1])
```

```python
# 정답과 예측 잘 맞나 비교
(yt == yp)
```

```python
matched = (yt == yp)
matched.sum()
```

```python
# 맞은 갯수 (정답 수)
correct = matched.sum()

# 전체 수
total = len(matched)

# 정확도
accuracy = correct / total
print(accuracy)
```

```python
print(f'정답수:{correct}, 전체개수:{total}, 정확도:{accuracy*100}%')
```

##벡터 값을 1차 함수로 변환 [0,1] 범위 제한

```python
# 변환 >> 정규화
(n1 - n1_min) / (n1_max - n1_min)
```
