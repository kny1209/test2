```python
# 라이브러리 임포트

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# 폰트 관련 용도
import matplotlib.font_manager as fm

# 나눔 고딕 폰트의 경로 명시
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_name = fm.FontProperties(fname=path, size=10).get_name()
```

```python
# 기본 폰트 설정
plt.rcParams['font.family'] = font_name

# 기본 폰트 사이즈 변경
plt.rcParams['font.size'] = 14

# 기본 그래프 사이즈 변경
plt.rcParams['figure.figsize'] = (6,6)

# 기본 그리드 표시
# 필요에 따라 설정할 때는, plt.grid()
plt.rcParams['axes.grid'] = True

# 마이너스 기호 정상 출력
plt.rcParams['axes.unicode_minus'] = False

# 넘파이 부동소수점 자릿수 표시
np.set_printoptions(suppress=True, precision=4)
```

```python
import torch

x1 = torch.ones(5)
print(x1.shape)
# x1: shape=[5] 가 되는 모든 값이 1인 텐서

print(x1)

x2 = x1.data.numpy()
print(x2)
```

```python
# x1 값 변경
x1[1] = -1

print(x1)
print(x2)
# x2의 값도 같이 변함
```

```python
# x1: tensor

x1 = torch.ones(5)

# x2: x1를 copy >> numpy
x2 = x1.data.numpy().copy()

x1[1] = -1

print(x1)
print(x2)
```

##합성함수를 파이썬으로 구현

```python
def f(x):
    return (2*x**2 + 2)
```

```python
# numpy 배열로 x 정의
np.arange(-2,2.1,0.25)
```

```python
x = np.arange(-2,2.1,0.25)
print(x)
```

```python
y = f(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# 세 가지 기본 함수의 정의


def f1(x):
    return(x**2)


def f2(x):
    return(x*2)


def f3(x):
    return(x+2)


# 합성 함수 만들기


x1 = f1(x)
x2 = f2(x1)
y = f3(x2)
```

```python
print(y)
```
