문제 2 (실습 문제 - 코드 작성)

그런 다음, nn.init.constant_를 사용하여 이 함수 l의 가중치(weight)는 모두 2.0으로, 편향(bias)은 0.5로 초기화하고, 초기화된 l.weight와 l.bias를 출력하시오.

문제 3 (실습 문제 - 코드 작성)

nn.Module을 상속받는 커스텀 모델 클래스 MyRegressionModel을 정의하시오.

[요구사항]

__init__에서 입력 차원 n_input과 출력 차원 n_output을 인자로 받아, nn.Linear 레이어를 self.linear_layer라는 이름으로 정의해야 합니다.

forward 메소드에서 입력 x가 self.linear_layer를 통과한 결과를 반환해야 합니다.

```python
import torch
import torch.nn as nn

class MyRegressionModel(nn.Module):
    def __init__(self, n_input, n_output):
        # TODO: 1. 부모 클래스의 __init__ 호출
        super(MyRegressionModel,self).__init__()

        # TODO: 2. self.linear_layer 라는 이름으로 nn.Linear 레이어 정의
        self.linear_layer = nn.Linear(n_input, n_output)

    def forward(self, x):
        # TODO: 3. linear_layer를 통과한 예측값 반환
        return self.linear_layer(x)

# --- 테스트 코드 (수정 불필요) ---
# 입력 특성 5개, 출력 특성 1개
net = MyRegressionModel(n_input=5, n_output=1)
print(net)

# (배치 크기 2, 특성 5)의 더미 입력으로 테스트
dummy_input = torch.randn(2, 5)
output = net(dummy_input)
print(f"\n입력 크기: {dummy_input.shape}")
print(f"출력 크기: {output.shape}")
```

문제 4 (실습 문제 - 코드 빈칸 채우기)

선형 회귀 모델의 학습을 위해 모델, 손실 함수, 옵티마이저를 정의하는 코드입니다.

3개의 빈칸 (# TODO: ...)을 채우시오.


```python
import torch
import torch.nn as nn
import torch.optim as optim

# --- 설정값 (수정 불필요) ---
n_input = 1  # 입력 특성 수 (예: 방 개수)
n_output = 1 # 출력 특성 수 (예: 가격)
learning_rate = 0.01

# --- 모델 클래스 (수정 불필요) ---
class Net(nn.Module):
    def __init__(self, n_input, n_output):
        super().__init__()
        self.l1 = nn.Linear(n_input, n_output)
    def forward(self, x):
        return self.l1(x)

# TODO: 1. Net 클래스를 사용하여 n_input, n_output에 맞는 모델 'net' 인스턴스 생성
net = Net(n_input, n_output)

# TODO: 2. 회귀 문제에 적합한 '평균 제곱 오차' 손실 함수 'criterion' 생성
criterion = nn.MSELoss()

# TODO: 3. 'SGD' 옵티마이저 'optimizer' 생성
# (net의 파라미터와 learning_rate를 인자로 전달)
optimizer = optim.SGD(net.parameters(), lr = learning_rate)


# --- 결과 확인 (수정 불필요) ---
print(f"모델:\n{net}")
print(f"\n손실 함수:\n{criterion}")
print(f"\n옵티마이저:\n{optimizer}")
```

문제 5 (실습 문제 - 코드 빈칸 채우기)

경사 하강법을 이용한 학습 루프(Training Loop)의 핵심 4단계입니다.

4개의 빈칸 (# TODO: ...)을 올바른 순서대로 채우시오.


```python
import torch
import torch.nn as nn
import torch.optim as optim

# --- 가상의 모델, 데이터, 손실함수, 옵티마이저 (수정 불필요) ---
net = nn.Linear(1, 1) # 모델
criterion = nn.MSELoss() # 손실 함수
optimizer = optim.SGD(net.parameters(), lr=0.01) # 옵티마이저
inputs = torch.randn(10, 1) # 입력 데이터
labels = torch.randn(10, 1) # 정답 데이터

# --- 학습 루프 1회 반복 ---

# TODO: 1. 예측 계산 (모델에 입력을 통과시킴)
outputs = net(inputs)

# TODO: 2. 손실 계산 (예측값과 정답 비교)
loss = criterion(outputs, labels)

# TODO: 3. 경사 초기화 (이전 반복의 경사를 지움)
optimizer.zero_grad()

# TODO: 4. 경사 계산 (손실에 대해 역전파 수행)
loss.backward()

# TODO: 5. 파라미터 수정 (옵티마이저 실행)
optimizer.step()


# --- 결과 확인 (수정 불필요) ---
print(f"계산된 손실 값: {loss.item():.4f}")
print("학습 루프 1회 실행 완료.")
# (참고: net.parameters()의 경사 값(grad)은 step() 이후 0으로 초기화되어야 함)
```
