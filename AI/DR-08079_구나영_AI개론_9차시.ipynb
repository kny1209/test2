문제 3 (실습 문제 - 코드 작성)

nn.Module을 상속받는 다중 분류(Multi-class Classification) 모델 클래스 MultiClassNet을 정의하시오.

[요구사항]

__init__: n_input (입력 특성 수), n_output (클래스 개수)를 인자로 받습니다.

__init__: self.linear라는 이름으로 nn.Linear 계층을 정의합니다.

(Softmax 계층은 정의하지 않습니다.)

forward: self.linear를 통과한 원본 출력(logits)을 그대로 반환합니다.

```python
import torch
import torch.nn as nn

class MultiClassNet(nn.Module):
    def __init__(self, n_input, n_output):
        # TODO: 1. 부모 클래스의 __init__ 호출
        super().__init__()

        # TODO: 2. self.linear 라는 이름으로 nn.Linear 계층 정의
        self.linear = nn.Linear(n_input, n_output)

    def forward(self, x):
        # TODO: 3. linear 계층을 통과한 원본 출력(logit) 반환
        return self.linear(x)

# --- 테스트 코드 (수정 불필요) ---
# 입력 특성 4개, 출력 클래스 3개
net = MultiClassNet(n_input=4, n_output=3)
print(net)

# (배치 크기 2, 특성 4)의 더미 입력으로 테스트
dummy_input = torch.randn(2, 4)
output = net(dummy_input)
print(f"\n입력 크기: {dummy_input.shape}")
print(f"출력 크기: {output.shape}")
```

문제 4 (실습 문제 - 코드 빈칸 채우기)

iris 데이터를 불러와 훈련/검증 데이터로 나누고 텐서로 변환하는 코드입니다.

nn.CrossEntropyLoss를 사용하기 위해 정답(label) 텐서의 데이터 타입을 올바르게 지정해야 합니다.

3개의 빈칸 (# TODO: ...)을 채우시오.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import torch

# --- 데이터 로드 및 분할 (수정 불필요) ---
iris = load_iris()
x_data, y_data = iris.data, iris.target
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.5, random_state=123)
# ------------------------------------

# TODO: 1. 훈련용 입력(x_train)을 'float' 타입 텐서 'inputs'로 변환
inputs = torch.tensor(x_train).float()

# TODO: 2. 훈련용 정답(y_train)을 'long' 타입 텐서 'labels'로 변환
labels = torch.tensor(y_train).long()

# TODO: 3. 검증용 정답(y_test)을 'long' 타입 텐서 'labels_test'로 변환
labels_test = torch.tensor(y_test).long()


# --- 결과 확인 (수정 불필요) ---
print(f"Inputs dtype: {inputs.dtype}")
print(f"Labels dtype: {labels.dtype}")
print(f"Labels_test dtype: {labels_test.dtype}")
```

문제 5 (실습 문제 - 코드 빈칸 채우기)

다중 분류 학습을 위해 모델, 손실 함수, 옵티마이저를 정의하는 코드입니다.

2개의 빈칸 (# TODO: ...)을 채우시오.


```python
import torch.optim as optim
import torch.nn as nn

# --- 설정값 (수정 불필요) ---
n_input = 4  # 입력 특성 수
n_output = 3 # 클래스 개수
learning_rate = 0.01

# (문제 3의 모델 클래스 사용)
class MultiClassNet(nn.Module):
    def __init__(self, n_input, n_output):
        super().__init__()
        self.linear = nn.Linear(n_input, n_output)
    def forward(self, x):
        return self.linear(x)

net = MultiClassNet(n_input, n_output)
# ----------------------------

# TODO: 1. 다중 분류에 적합한 '교차 엔트로피' 손실 함수 'criterion' 생성
criterion = nn.CrossEntropyLoss()

# TODO: 2. 'SGD' 옵티마이저 'optimizer' 생성
# (net의 파라미터와 learning_rate를 인자로 전달)
optimizer = optim.SGD(net.parameters(), lr = learning_rate)


# --- 결과 확인 (수정 불필요) ---
print(f"모델:\n{net}")
print(f"\n손실 함수:\n{criterion}")
print(f"\n옵티마이저:\n{optimizer}")
```

문제 6 (실습 문제 - 코드 빈칸 채우기)

모델의 원본 출력(logits) 텐서 outputs가 주어졌을 때, (1) 최종 예측 클래스 predicted와 (2) 정확도 accuracy를 계산하는 코드입니다.

3개의 빈칸 (# TODO: ...)을 채우시오.

```python
import torch

# --- 가상의 모델 출력값과 정답 (수정 불필요) ---
# outputs: 모델이 출력한 원본 Logits (Softmax 적용 전)
# (배치 크기 3, 클래스 3)
outputs = torch.tensor([[1.0, 3.0, 0.5],
                        [2.0, 1.0, 0.0],
                        [0.0, 0.0, 5.0]])
# labels: 실제 정답 (long 타입)
labels = torch.tensor([1, 0, 2])
# ------------------------------------------------

# TODO: 1. 'outputs' 텐서의 1번 차원(행별)에서 최댓값의 '인덱스' 'predicted' 추출 [cite: 375, 381]
# (힌트: torch.max()는 2개의 값을 반환합니다)
predicted = torch.max(outputs,1)[1]

# TODO: 2. 'predicted'와 'labels'가 일치하는 건수 'correct' 계산
correct = (predicted == labels).sum()

# TODO: 3. 'correct'를 전체 건수(len(labels))로 나누어 'accuracy' 계산
accuracy = correct / len(labels)


# --- 결과 확인 (수정 불필요) ---
# (outputs의 행별 최댓값 인덱스: [1, 0, 2])
# (labels: [1, 0, 2])
# (결과: 3 / 3 = 1.0)
print(f"모델 출력 (Logits):\n{outputs}")
print(f"예측 클래스 (Indices): {predicted}")
print(f"정답 클래스 (Labels): {labels}")
print(f"맞춘 건수: {correct}")
print(f"정확도: {accuracy:.2f}")
```
