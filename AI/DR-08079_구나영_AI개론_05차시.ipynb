문제 3(실습 코드 작성)

파이토치의 nn.Module을 상속받아 선형 회귀(Linear Regression) 모델 역할을 하는 클래스 LinearModel을 정의하시오.

요구사항:

1. 생성자(__init__)에서 입력 특성 수(input_dim)와 출력 특성 수(output_dim)를 인자로 받아, nn.Linear 레이어를 self.linear라는 이름으로 초기화해야 합니다.
2. 순전파 메소드(forward)는 입력 텐서 x를 받아 self.linear 레이어를 통과시킨 결과를 반환해야 합니다. (활성화 함수 없음)

```python
import torch
import torch.nn as nn

class LinearModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearModel, self).__init__()
        # TODO: 1. nn.Linear 레이어를 self.linear로 정의
        self.linear = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        # TODO: 2. self.linear 레이어를 통과한 출력을 반환
        return self.linear(x)

# --- 테스트 코드 (수정 불필요) ---
# 입력 특성 10개, 출력 특성 1개 (회귀 문제)
model = LinearModel(input_dim=10, output_dim=1)
print(model)

# 테스트 입력 데이터 (배치 크기 2, 특성 10개)
test_input = torch.randn(2, 10)
output = model(test_input)

print(f"\n입력 데이터 크기: {test_input.shape}")
print(f"출력 데이터 크기: {output.shape}")
```

문제 4(실습 코드 작성)

문제 3에서 정의한 LinearModel 클래스를 사용하여, 주어진 데이터 X와 Y로 선형 회귀 모델을 학습하는 전체 과정을 완성하시오.

요구사항:

1. 모델을 input_dim=1, output_dim=1로 생성하시오.
2. 손실 함수는 nn.MSELoss() (평균 제곱 오차)를 사용하시오.
3. 옵티마이저는 optim.SGD를 사용하고, 모델의 파라미터(model.parameters())와 학습률(lr=0.01)을 지정하시오.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# --- 문제 3에서 정의한 모델 클래스 (수정 불필요) ---
class LinearModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearModel, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim)
    def forward(self, x):
        return self.linear(x)

# --- 데이터 (수정 불필요) ---
# y = 2x + 1 에 노이즈 추가
X = torch.randn(100, 1)
Y = 2 * X + 1 + 0.1 * torch.randn(100, 1)

# --- 학습 파라미터 (수정 불필요) ---
learning_rate = 0.01
epochs = 100

# TODO: 1. 모델 정의 (input_dim=1, output_dim=1)
model = LinearModel(input_dim=1, output_dim=1)

# TODO: 2. 손실 함수 정의 (MSELoss)
criterion = nn.MSELoss()

# TODO: 3. 옵티마이저 정의 (SGD)
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

# --- 학습 루프 (수정 불필요) ---
for epoch in range(epochs):
    # 1. 예측
    Yp = model(X)

    # 2. 손실 계산
    loss = criterion(Yp, Y)

    # 3. 경사 초기화
    optimizer.zero_grad()

    # 4. 역전파 (경사 계산)
    loss.backward()

    # 5. 파라미터 업데이트
    optimizer.step()

    if (epoch + 1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# --- 학습된 파라미터 확인 (수정 불필요) ---
# W는 2.0에, B는 1.0에 가까워야 함
[W, B] = model.parameters()
print(f"\n학습된 가중치(W): {W.item():.4f}")
print(f"학습된 편향(B): {B.item():.4f}")
```

문제 5 (실습 문제 - 코드 작성)
- nn.Sequential을 사용하지 않고 __init__과 forward 메소드만으로 2개의 은닉층을 가진 분류(Classification) 모델 클래스 ClassifierModel을 정의하시오.

구조:
- 입력층 (nn.Linear): input_dim -> 64
- 활성화 함수 (nn.ReLU)
- 은닉층 1 (nn.Linear): 64 -> 32
- 활성화 함수 (nn.ReLU)
- 출력층 (nn.Linear): 32 -> output_dim (예: 1)
- 출력 활성화 함수 (nn.Sigmoid): 분류 문제이므로 forward의 마지막에 시그모이드 함수를 적용해야 함.

```python
import torch
import torch.nn as nn

class ClassifierModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(ClassifierModel, self).__init__()
        # TODO: 1. 필요한 레이어들(Linear, ReLU, Sigmoid)을 정의
        # (nn.ReLU와 nn.Sigmoid는 인스턴스로 생성해두는 것이 일반적임)
        self.layer1 = nn.Linear(input_dim, 64)
        self.relu1 = nn.ReLU()
        self.layer2 = nn.Linear(64, 32)
        self.relu2 = nn.ReLU()
        self.layer3 = nn.Linear(32, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # TODO: 2. 정의된 레이어들을 순서대로 통과시키는 순전파 로직 구현
        # (Input -> Linear1 -> ReLU1 -> Linear2 -> ReLU2 -> Linear3 -> Sigmoid -> Output)
        x = self.layer1(x)
        x = self.relu1(x)
        x = self.layer2(x)
        x = self.relu2(x)
        x = self.layer3(x)
        x = self.sigmoid(x)
        return x

# --- 테스트 코드 (수정 불필요) ---
# 입력 특성 20개, 출력 특성 1개 (이진 분류)
model_cls = ClassifierModel(input_dim=20, output_dim=1)
print(model_cls)

# 테스트 입력 데이터 (배치 크기 3, 특성 20개)
test_input_cls = torch.randn(3, 20)
output_cls = model_cls(test_input_cls)

print(f"\n입력 데이터 크기: {test_input_cls.shape}")
print(f"출력 데이터 크기: {output_cls.shape}")
print(f"출력 값 (Sigmoid 적용): {output_cls.data}") # 0과 1 사이의 값이 출력되어야 함
```

문제 6 (실습 문제 - 코드 작성)

이번에는 nn.Sequential을 사용하여 문제 5와 동일한 구조의 분류 모델 ClassifierModelSeq를 정의하시오.

nn.Sequential 컨테이너 내부에 레이어들을 순서대로 추가해야 합니다.

구조: Linear -> ReLU -> Linear -> ReLU -> Linear -> Sigmoid

```python
import torch
import torch.nn as nn

class ClassifierModelSeq(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(ClassifierModelSeq, self).__init__()

        # TODO: 1. self.network 라는 이름으로 nn.Sequential 모델을 정의
        self.network = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, output_dim),
            nn.Sigmoid()
        )

    def forward(self, x):
        # TODO: 2. self.network를 통과한 출력을 반환
        return self.network(x)

# --- 테스트 코드 (수정 불필요) ---
# 입력 특성 20개, 출력 특성 1개 (이진 분류)
model_seq = ClassifierModelSeq(input_dim=20, output_dim=1)
print(model_seq)

# 테스트 입력 데이터 (배치 크기 3, 특성 20개)
test_input_seq = torch.randn(3, 20)
output_seq = model_seq(test_input_seq)

print(f"\n입력 데이터 크기: {test_input_seq.shape}")
print(f"출력 데이터 크기: {output_seq.shape}")
print(f"출력 값 (Sigmoid 적용): {output_seq.data}") # 0과 1 사이의 값이 출력되어야 함
```
