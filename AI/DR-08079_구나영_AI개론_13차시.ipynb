문제 3 (실습 문제 - 코드 빈칸 채우기)

torchvision.models에서 VGG16 모델을 불러온 뒤, 전이 학습(Transfer Learning)을 위해 사전 학습된 가중치(Convolutional Base)를 모두 고정(freeze)하는 코드입니다.

2개의 빈칸 (# TODO: ...)을 채워 모든 파라미터의 requires_grad를 False로 설정하시오.

문제 4 (실습 문제 - 코드 빈칸 채우기)

VGG16의 분류기(classifier)는 7개의 계층으로 구성되어 있으며, 그 중 마지막(인덱스 6) 계층을 교체해야 합니다.

3개의 빈칸 (# TODO: ...)을 채우시오.

```python
import torch.nn as nn
import torchvision.models as models

# --- 사전 학습된 VGG16 모델 로드 (수정 불필요) ---
model_vgg = models.vgg16(pretrained=True)
print(f"원본 분류기 마지막 레이어:\n{model_vgg.classifier[6]}\n")
# ------------------------------------------------

# TODO: 1. 마지막 분류기의 입력 특성 수(in_features) 가져오기
# (힌트: model_vgg.classifier[6].in_features)
n_features = model_vgg.classifier[6].in_features

# TODO: 2. 새로운 출력 클래스 수
n_classes = 2

# TODO: 3. n_features를 입력받아 n_classes를 출력하는 새로운 nn.Linear 계층으로
# model_vgg.classifier의 6번 인덱스를 교체
model_vgg.classifier[6] = nn.Linear(n_features, n_classes)


# --- 결과 확인 (수정 불필요) ---
print(f"변경된 분류기 마지막 레이어:\n{model_vgg.classifier[6]}")
```

문제 5 (실습 문제 - 코드 빈칸 채우기)

CIFAR-10용 CNN 모델 Net의 forward 메소드입니다. Conv2d와 MaxPool2d를 통과한 4D 텐서 x를 Linear 계층에 입력하기 위해 1D 벡터로 평탄화(flatten)하는 과정입니다.

2개의 빈칸 (# TODO: ...)을 채워 [배치크기, 16, 5, 5] 형태의 텐서를 [배치크기, 400] 형태로 변환하시오.

```python
import torch
import torch.nn as nn

# --- CNN 모델 정의 (일부 수정 불필요) ---
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)

        # TODO: 1. 16*5*5 크기의 1D 벡터를 입력받는 'fc1' 정의
        self.fc1 = nn.Linear(400, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x): # x의 Shape: [N, 3, 32, 32]
        x = self.pool(F.relu(self.conv1(x))) # [N, 6, 14, 14]
        x = self.pool(F.relu(self.conv2(x))) # [N, 16, 5, 5]

        # TODO: 2. x 텐서를 fc1에 입력하기 위해 1D 벡터로 평탄화(flatten)
        # (힌트: 배치 크기(N)는 유지 -> [N, 16*5*5])
        x = x.view(-1, 400)

        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
# ------------------------------------------

# --- 테스트 코드 (수정 불필요) ---
import torch.nn.functional as F
net = Net()
print(net)

# (배치 4, 채널 3, 높이 32, 너비 32)의 가상 입력
dummy_images = torch.randn(4, 3, 32, 32)
output = net(dummy_images)

print(f"\n입력 Shape: {dummy_images.shape}")
print(f"최종 출력 Shape: {output.shape}") # [4, 10]
```

문제 6 (실습 문제 - 코드 작성)

CIFAR-10 데이터셋을 위한 전처리(transform) 파이프라인과 DataLoader를 생성하는 코드입니다.

[요구사항]

transform: transforms.Compose를 사용하여 2가지 변환을 순서대로 정의합니다.

- transforms.ToTensor()
- transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)) (RGB 3채널이므로 3개의 값 사용)
- trainloader: trainset을 batch_size=4, shuffle=True로 설정하여 생성합니다.


```python
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.utils.data import DataLoader

# TODO: 1. ToTensor와 Normalize를 포함하는 'transform' 파이프라인 정의
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# --- 데이터셋 로드 (수정 불필요) ---
trainset = datasets.CIFAR10(root='./data', train=True,
                           download=True, transform=transform)
# -----------------------------------

batch_size = 4

# TODO: 2. 훈련용 'trainloader' 생성 (batch_size=4, shuffle=True)
trainloader = DataLoader(trainset, batch_size=batch_size, shuffle = True)


# --- 결과 확인 (수정 불필요) ---
print(transform)
print(f"\nTrain DataLoader: {trainloader}")
print(f"Batch Size: {trainloader.batch_size}")
print(f"Shuffle: {trainloader.sampler}") # Shuffle=True면 RandomSampler가 나옴
```
